```python
#remove unnecessary columns
df = df[['ID', 'Severity', 'Start_Time', 'End_Time', 'Start_Lat', 'Start_Lng',
    'End_Lat', 'End_Lng', 'Distance(mi)', 
    #'Description', 'Number', 'Street','Side', 
    'City', 'County', 'State', 'Zipcode', 
    #'Country', 'Timezone','Airport_Code', 
    'Weather_Timestamp', 'Temperature(F)', 
    #'Wind_Chill(F)',
       'Humidity(%)', 
    #'Pressure(in)', 
    'Visibility(mi)', 
    #'Wind_Direction',
       'Wind_Speed(mph)', 'Precipitation(in)', 'Weather_Condition', 'Amenity',
       'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway',
       'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal',
       'Turning_Loop', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight',
       'Astronomical_Twilight']].copy()
```

```python
#convert time into datetime format
df['End_Time'] = pd.to_datetime(df['End_Time'])
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
```

```python
#rename columns
#df.rename(columns={'ID':'ID2','City':'City2'})
```

```python
#check any missing values in the columns
df.isna().sum()
```

```python
#remove duplicate records
df = df.loc[~df.duplicated(subset=['Start_Time','End_Time','Start_Lat', 'Start_Lng','End_Lat', 'End_Lng'])].reset_index(drop=True)
```

```python
df.shape
```

```python
#plot the highest number of accident by city
ax = df['City'].value_counts().head(10).plot(kind='barh', title='Highest Accident City')
ax.set_xlabel('Number')
ax.set_ylabel('City')
```

```python
#plot histrogram sample
df['Severity'].plot(kind='hist')
```

```python
#plot scatter plot sample
df.plot(kind='scatter', x='Visibility(mi)', y='Precipitation(in)')
```

```python
#plot different scatter plot and determine relationship
sns.pairplot(df, vars=['Severity', 'Temperature(F)', 'Humidity(%)',  'Visibility(mi)', 'Wind_Speed(mph)'])
plt.show()
```

```python
#remove empty value
df = df.dropna().reset_index(drop=True)
```

```python
#remove impossible temperature
df = df[df['Temperature(F)'] < 122].reset_index(drop=True)
```

```python
#remove impossible windspeed
df = df[(df['Wind_Speed(mph)'] < 200) & (df['Wind_Speed(mph)'] >= 0)].reset_index(drop=True)
```

```python
#remove weather timestamp columns
df= df.drop(columns='Weather_Timestamp')
```

```python
#save into csv
df.to_csv('US_Accidents_cleaned.csv', encoding='utf-8')
```
